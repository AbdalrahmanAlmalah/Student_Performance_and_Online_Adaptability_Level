```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Student Performance.csv')
X = df.loc[: , df.columns != 'writing score']
y = df['writing score']
df
```

```python
l1 = LabelEncoder()
l1.fit(X[['gender']])
df['gender_res'] = l1.transform(df[['gender']])
df = df.drop('gender', axis=1)
df
```

```python
l2 = LabelEncoder()
l2.fit(X[['lunch']])
df['lunch_res'] = l2.transform(df[['lunch']])
df = df.drop('lunch', axis=1)
df
```

```python
l3 = LabelEncoder()
l3.fit(X[['test preparation course']])
df['test preparation course_res'] = l3.transform(df[['test preparation course']])
df = df.drop('test preparation course', axis=1)
df
```

```python
e1 = OneHotEncoder(sparse=False)
res1= e1.fit_transform(df[['race/ethnicity']])
df = pd.get_dummies(df, columns=['race/ethnicity'])
df
```

```python
e2 = OneHotEncoder(sparse=False)
res2= e2.fit_transform(df[['parental level of education']])
df = pd.get_dummies(df, columns=['parental level of education'])
df
```

```python
X = df.loc[: , df.columns != 'writing score']
y = df['writing score']
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test, y_train , y_test = train_test_split(X, y, test_size = 0.2)
```

# Multiple Linear

```python
from sklearn import linear_model

clf = linear_model.LinearRegression()
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)
y_predict
```

```python
print('multiple linear model coeff (w)', clf.coef_)
print('multiple linear model intercept (b)' , clf.intercept_)
print('R-squared score (training)', clf.score(X_train, y_train))
print('R-squared score (test)', clf.score(X_test,y_test))
```

```python
colors = []
for y1, y2 in zip(y_predict, y_test):
    if(abs(y2-y1)<5):
        colors.append('green')
    elif(abs(y2-y1)<10):
        colors.append('orange')
    else:
        colors.append('red')
```

```python
plt.figure()
plt.scatter(y_test,y_predict, c=colors)
plt.xlabel('Expected values')
plt.ylabel('Prediction')
plt.title('Multiple Linear')
plt.show()
```

# Ridge:

```python
from sklearn.linear_model import Ridge

linridge = Ridge(alpha=20.0).fit(X_train, y_train)
y_predict = linridge.predict(X_test)
y_predict
```

```python
print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
```

```python
colors = []
for y1, y2 in zip(y_predict, y_test):
    if(abs(y2-y1)<5):
        colors.append('green')
    elif(abs(y2-y1)<10):
        colors.append('orange')
    else:
        colors.append('red')
```

```python
plt.figure()
plt.scatter(y_test,y_predict, c=colors)
plt.xlabel('Expected values')
plt.ylabel('Prediction')
plt.title('Ridge')
plt.show()
```

# Polynomial:

```python

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Transform the input feature into polynomial features
degree = 2  # Set the degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
y_pred_train = model.predict(X_train_poly)
y_pred_test = model.predict(X_test_poly)
```

```python
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
print("Train Mean Squared Error:", mse_train)
print("Test Mean Squared Error:", mse_test)
```

```python
colors = []
for y1, y2 in zip(y_pred_test, y_test):
    if(abs(y2-y1)<5):
        colors.append('green')
    elif(abs(y2-y1)<10):
        colors.append('orange')
    else:
        colors.append('red')
```

```python
plt.figure()
plt.scatter(y_test, y_pred_test, c=colors)
plt.xlabel('Expected values')
plt.ylabel('Prediction')
plt.title('Polynomial')
plt.show()
```
